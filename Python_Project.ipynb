```python
import seaborn as sns
import pandas as pd
```

```python
#Loading the 'mpg' dataset
sns.set_theme()
mpg_data = sns.load_dataset('mpg')
```

```python
#Loading the first 5 rows of 'mpg_data'
mpg_data.head()
```

```python
#Importing the statement that allows us to use pyplot module from matplotlib
import matplotlib.pyplot as plt
```

```python
#Create a visual showing the relationship between horsepower and fuel economy
sns.relplot(
    data= mpg_data,
    x="horsepower", y="mpg"
)
#Titles and lables 
plt.title("Relationship Between Horsepower and Fuel Economy")
plt.xlabel("Horsepower")
plt.ylabel("Miles Per Gallon (MPG)")
    
```

```python
#Create a table summarizing fuel economy by country of origin
fe_summary = mpg_data.groupby('origin')['mpg'].agg(['mean', 'std', 'min', 'max'])
fe_summary.columns = ['Mean MPG', 'Standard Deviation', 'Minimum MPG', 'Maximum MPG']
fe_summary.head()
```

```python
#Create a visual showing how fuel economy has changed over time
fuel_economy_by_year = mpg_data.groupby('model_year')['mpg'].mean().reset_index()

sns.lineplot(data=fuel_economy_by_year, x='model_year', y='mpg', marker='o')

#Titles and labels
plt.title('Change in Fuel Economy Over Time')
plt.xlabel('Model Year')
plt.ylabel('Mean Miles per Gallon (MPG)')
```

```python
#Create two other visuals or tables related to fuel economy

# Create a boxplot to show the distribution of fuel economy by country of origin
sns.boxplot(data=mpg_data, x='origin', y='mpg', hue='origin', palette='viridis', dodge=False)

# Add titles and labels
plt.title('Distribution of Fuel Economy by Country of Origin')
plt.xlabel('Country of Origin')
plt.ylabel('Miles per Gallon (MPG)')
```

```python
fuel_economy_by_cylinders = mpg_data.groupby('cylinders')['mpg'].mean().reset_index()

# Set the Seaborn theme
sns.set_theme()

# Create a bar plot to show fuel economy by number of cylinders
sns.barplot(data=fuel_economy_by_cylinders, x='cylinders', y='mpg', hue='cylinders', palette='viridis', dodge=False)
# Add titles and labels
plt.title('Fuel Economy by Number of Cylinders')
plt.xlabel('Number of Cylinders')
plt.ylabel('Mean Miles per Gallon (MPG)')
```
